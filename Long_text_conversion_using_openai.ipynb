```python
import pandas as pd
import openai

```

```python
df=pd.read_csv('Book1.csv')
```

```python
df.head()
```

```python
# Set your OpenAI API key
openai.api_key = ''

# Function to generate long item description using OpenAI GPT-3 API
def generate_long_description(short_description):
    prompt = f"Use {short_description} to generate text only in less than 70 words with no numerical values and special characters in generated text. Please do not hallucinate."
    response = openai.Completion.create(
        model="gpt-3.5-turbo-instruct",
        #engine="davinci-002",
        prompt=prompt,
        temperature=0,
        max_tokens=150,
        n=1,
        stop=None,
    )
    return response['choices'][0]['text'].strip()

# Sample data
#data = {'short_description': ['MIRROR LOVE BIRD T-LIGHT HOLDER', 'GLITTER SNOW PEAR TREE DECORATION', 'GINGHAM OVEN GLOVE RED HEART ']}
#df = pd.DataFrame(data)


```

```python

# Generate long item descriptions and add to dataframe
df['Details'] = df['Description'].apply(generate_long_description)

```
